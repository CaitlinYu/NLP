## Advantages and disadvantages of classic machine learning models

# Supervised ML

### Model1: Linear Regression

### Advantages:
1. 模型简单，训练速度快
2. 容易理解，能很清晰的看到模型的参数

### Disadvantages:
1. 不能拟合非线性关系的数据，很有局限性
2. 不适合skewed dataset

### Model2: KNN

### Advantages:
1. 原理简单易懂，应用起来方便
2. 调参容易，只有两个参数，distance和k

### Disadvantages:
1. 预测的时候速度慢，因为每次预测都要计算这个点和相邻点的距离。
2. 也很耗内存，memory consumption，因为要存所有的数据
3. 需要做数据预处理，normalization，否则会让数值大的特征主宰预测结果
4. 不适用于维度很高的数据，因为当维度过高的时候，点之间的距离就都差不多了，不具有区分性。所以得先做数据降维
5. Outlier对结果影响很大

### Model3: Logistic Regression
是分类模型

### Advantages:
1. 模型简单，高效，训练速度快
2. 耗内存少

### Disadvantages:
1. 不能结局非线性分类问题
2. 需要做特诊工程，选择少量的重要的特征，避免overfitting

### Model4: Support Vector Machine
是分类模型

### Advantages:
1. 不需要迭代iteration就能找到最优解
2. 对不平衡数据比较适合
3. 能解决比较复杂的分类问题，使用合适的kernal function核函数

### Disadvantages:
1. 数据集比较大时，训练速度慢
2. 寻找适合的核函数不容易
3. 模型不容易理解和可视化

### Model5: Decision Tree
是分类模型

### Advantages:
1. 能很快找出最显著特征
2. 方便理解，有理有据
3. 可视化比较好，用graph_viz的包

### Disadvantages:
1. 容易overfit 
2. 不太稳定，当数据增加或减少一些时
3. 准确度没有别的ML模型高。但是可以用随机森林做替代。

### Model6: Random Forest
是分类模型

### Advantages:
1. 泛化能力强，因为每次都随机选择一部分特征做决策树，这样避免了overfitting.
2. 灵活性高，准确度高
3. 比较稳定，当数据增减的时候

### Disadvantages:
1. 不容易理解和可视化
2. 训练耗时耗内存

### Model7: Neural Network


### Advantages:
1. Powerful 函数拟合能力强
2. Generalization 泛化能力很强
3. Versatile 多样化，可变性高

### Disadvantages:
1. Data feeding 耗数据
2. Black box 不可解释

# Unsupervised ML (Mostly Clustering)

### Model8: K-means Clustering
是聚类模型

### Advantages:
1. 比较简单，容易应用
3. 一定会收敛
4. 对新的数据能很快适应

### Disadvantages:
1. 需要手动选择k
2. 对于初始点的选择有依赖，会有不同的运行时间，具有随机性
3. 需要先向量化
4. 训练时间比较长
5. 有些分布不适用这个方法，例如环状分布的数据集
