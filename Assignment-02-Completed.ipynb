````python
distance_list = list()
for i in soup.find_all(re.compile(r'td|th')):
    #if ('线相邻站间距信息统计表' in i.get_text()) & ('\n' not in i.get_text()):
    #    distance_list.append(i.get_text())
    if '――' in i.get_text():
        distance_list.append(i.get_text())
    elif i.get_text().isdigit():
        distance_list.append(i.get_text())

print(distance_list)
````

````python
def get_geo_distance(station1, station2):
    key_name1 = station1 + '――' + station2
    key_name2 = station2 + '――' + station1
    for i in range(len(distance_list)):
        if (distance_list[i] == key_name1) | (distance_list[i] == key_name2):
            return int(distance_list[i+1])
        else: continue
````

##### 3. Build the search agent

> Build the search agent based on the graph we build.

for example, when you run: 

```python
>>> search('奥体中心', '天安门') 
```
you need get the result: 

奥体中心-> A -> B -> C -> ... -> 天安门



````python
def bfs(graph, start):
    """
    breath first search
    """
    visited = [start]
    
    seen = set()
    
    while visited:
        frontier = visited.pop() #
        
        if frontier in seen: continue
        
        for successor in graph[frontier]:
            if successor in seen: continue
            print(successor)
            
            #visited = visited + [successor] # 我们每次扩展都扩展最新发现的点 -> depth first
            visited = [successor] + visited # 我们每次扩展都先考虑已经发现的 老的点 -> breath first
            
            # 所以说，这个扩展顺序其实是决定了我们的深度优先还是广度优先
    
        seen.add(frontier)
    
    return seen
````

````python
def search(start, destination, connection_graph):
    pathes = [[start]]
    
    visited = set()
    
    while pathes: # if we find existing pathes
        path = pathes.pop(0)
        frontier = path[-1]
        
        if frontier in visited: continue
            
        successors = connection_graph[frontier]
        #print(connection_graph)
        
        for city in successors:
            if city in path: continue  # eliminate loop
                
            new_path = path + [city]
            
            pathes.append(new_path)
            
            if city == destination: return new_path
        
        visited.add(frontier)
        
        #pathes = sort_candidate(pathes) # 我们可以加一个排序函数 对我们的搜索策略进行控制
    print(pathes)
````

````python
def transfer_count(path):
    lines = list()
    common = list()
    transfer_count = 0
    for i in path:
        lines.append(get_line_info(i))
    
    for i in range(len(lines)):
        common.extend(list(set(lines[i]) & set(lines[i-1])))

    for j in range(len(common)-1):
        if common[j+1] != common[j]:
            transfer_count += 1
        else: continue
    return transfer_count
````

````python
def get_path_distance(path):
    distance = 0

    for i in range(len(path)-1):
        if get_geo_distance(path[i], path[i+1]) != None:
            distance += get_geo_distance(path[i], path[i+1])
        else: continue
    #print(path, distance) # 显示每个路径的距离
    return distance
````

````python
def advanced_search(start, destination, connection_graph, strategy_name):
    pathes = [[start]]
    
    visited = set()
    
    while pathes: # if we find existing pathes
        path = pathes.pop(0)
        frontier = path[-1]
        
        if frontier in visited: continue
            
        successors = connection_graph[frontier]
        
        for city in successors:
            if city in path: continue  # eliminate loop
                
            new_path = path + [city] # depth first
            
            pathes.append(new_path)
            
            if city == destination: return new_path
        
        visited.add(frontier)
        
        pathes = strategy(pathes, strategy_name) # 我们可以加一个排序函数 对我们的搜索策略进行控制
    print(pathes)
````

````python
def get_comprehensive_score(path):
    # 通过计算路径的distance和transfer_count得到一个综合得分。
    # 如果能做到所有数据normalize就更好了。
    # 但是因为这个程序在search的过程中，每次都只能拿到一个path的两项得分，所以normalize起来比较复杂。
    # 简化一下，每个数值乘以一个系数，再加总。
    score = get_path_distance(path) * 0.00001 + transfer_count(path) * 0.1 + len(path) * 0.01
    #print(get_path_distance(path), transfer_count(path), len(path), score)
    return score
````

````python
def strategy(pathes, strategy_name):
    if len(pathes) <= 1:
        return pathes
    else:
        if strategy_name == 'shortest_path_priority':
            return sorted(pathes, key=get_path_distance)
        elif strategy_name == 'min_transfer_priority':
            return sorted(pathes, key=transfer_count)
        elif strategy_name == 'comprehensive_priority':
            return sorted(pathes, key=get_comprehensive_score)
        else:
            return pathes
````
