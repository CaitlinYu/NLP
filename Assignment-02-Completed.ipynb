## Change loss function to:
$Loss = \frac{1}{n} sum(|{y_i - \hat{y_i}}|)$

````python
def loss(y, y_hat):
    return sum(abs(y_i - y_hat_i) for y_i, y_hat_i in zip(list(y), list(y_hat))) / len(list(y))
````

###### 3.Gradient Descent to get optimal *k* and *b*

$$ Loss = \frac{1}{n} sum(|{y_i - \hat{y_i}}|)$$
When $\hat{y_i} < y_i$:
$$ \frac{\partial{loss}}{\partial{k}} = -\frac{1}{n}\sum(x_i)$$
$$ \frac{\partial{loss}}{\partial{b}} = -\frac{1}{n}\sum(1)$$
When $\hat{y_i} > y_i$:
$$ \frac{\partial{loss}}{\partial{k}} = \frac{1}{n}\sum(x_i)$$
$$ \frac{\partial{loss}}{\partial{b}} = \frac{1}{n}\sum(1)$$

````python
def partial_k(x, y, y_hat):
    n = len(y)
    gradient = 0
    for x_i, y_i, y_hat_i in zip(list(x), list(y), list(y_hat)):
        if y_hat_i < y_i:
            gradient += (-x_i)
        else:
            gradient += x_i
    return 1/n * gradient

def partial_b(x, y, y_hat):
    n = len(y)
    gradient = 0
    for y_i, y_hat_i in zip(list(y), list(y_hat)):
        if y_hat_i < y_i:
            gradient += -1
        else:
            gradient += 1
    return 1/n * gradient
````

````python
def bfs(graph, start):
    """
    breath first search
    """
    visited = [start]
    
    seen = set()
    
    while visited:
        frontier = visited.pop() #
        
        if frontier in seen: continue
        
        for successor in graph[frontier]:
            if successor in seen: continue
            print(successor)
            
            #visited = visited + [successor] #  -> depth first
            visited = [successor] + visited #   -> breath first
            
            # 
    
        seen.add(frontier)
    
    return seen
````

````python
def search(start, destination, connection_graph):
    pathes = [[start]]
    
    visited = set()
    
    while pathes: # if we find existing pathes
        path = pathes.pop(0)
        frontier = path[-1]
        
        if frontier in visited: continue
            
        successors = connection_graph[frontier]
        #print(connection_graph)
        
        for city in successors:
            if city in path: continue  # eliminate loop
                
            new_path = path + [city]
            
            pathes.append(new_path)
            
            if city == destination: return new_path
        
        visited.add(frontier)
        
        #pathes = sort_candidate(pathes) #  
    print(pathes)
````

````python
def transfer_count(path):
    lines = list()
    common = list()
    transfer_count = 0
    for i in path:
        lines.append(get_line_info(i))
    
    for i in range(len(lines)):
        common.extend(list(set(lines[i]) & set(lines[i-1])))

    for j in range(len(common)-1):
        if common[j+1] != common[j]:
            transfer_count += 1
        else: continue
    return transfer_count
````

````python
def strategy(pathes, strategy_name):
    if len(pathes) <= 1:
        return pathes
    else:
        if strategy_name == 'shortest_path_priority':
            return sorted(pathes, key=get_path_distance)
        elif strategy_name == 'min_transfer_priority':
            return sorted(pathes, key=transfer_count)
        elif strategy_name == 'comprehensive_priority':
            return sorted(pathes, key=get_comprehensive_score)
        else:
            return pathes
````
