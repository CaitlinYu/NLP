#### 1. Summarize the reasons of overfitting and underfitting. Put them in github repository.


Overfitting reasons:
1. Data is insufficient, in terms of number of data points or data granularity or both. If model does not have enough data to learn from, the model will not be able to generalize well from training data to unseen data, therefore overfits.
2. Noise interferes with signal in the data will cause overfitting. 
3. The algorithm is too complex or flexible (e.g. it has too many inputs features or it is not properly regularized), it can end up "memorizing the noise" instead of finding the signal.

--> Too complex model tends to have high variance and low bias.

Underfitting reasons:
1. Model is too simple (e.g. it has too few inputs features or regularized too much) which makes it inflexible in learning from the dataset.

--> Too simple model tends to have high bias and low variance.

#### 3. Writing down three sceneries that machine learning has been used now.


Below machine learning scenerios have been applied in banking and insurance industry:
1. Chatbots/AI assistants: Responding to internal agent inquiries and providing guidance on business protocols.
2. Driver performance monitoring: Machine learning algorithms are being applied to client data to help inform the development of products for insurance clients.
3. Insurance market analytics: Machine learning algorithms are being applied to interpret driver data in an effort to monitor market trends and identify business opportunities.
4. JPMorgan Chase and their COiN: In order to automate the daily routine and cut down the time needed to analyze the business correspondence, JPMorgan Chase has developed a proprietary ML algorithm called Contract Intelligence or COiN. It is now used to analyze the documentation and extract the important information from it. Applying this tool enabled the bank to process 12,000 credit agreements in several seconds, instead of 360,000 man-hours.
5. Credit decision: Banks are using machine learning to determine which applicants are higher default risks and which applicants are more credit-worthy, even without an extensive credit history.
