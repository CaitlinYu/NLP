#### 5. What's the Language Model;

Ans: Language Model refers to the model that calculate the proability of a sentence coming up in a paragraph, by looking into the probability of words appear. So as to make sense of which sentence is more reasonable and rational.
$$ language\_model(String) = Probability(String) \in (0, 1) $$
$$ Pro(w_1 w_2 w_3 w_4) = Pr(w_1 | w_2 w_3 w_ 4) * P(w2 | w_3 w_4) * Pr(w_3 | w_4) * Pr(w_4)$$ 
$$ Pro(w_1 w_2 w_3 w_4) \sim Pr(w_1 | w_2 ) * P(w2 | w_3 ) * Pr(w_3 | w_4) * Pr(w_4)$$

#### 7. What's the 1-gram language model;

Ans: 1-gram language model calculates the sentence probability based on the probability that very single word appears in a paragraph. The counting is done compared to a collection of single words.

#### 8. What's the disadvantages and advantages of 1-gram language model;

Ans: Disadvantages:
1. Breaking phrases into words loses meanings of the phrase sometimes.
2. Unable to catch words with length more than one.
3. Lack of explicit dependency range.

Advantages:
1. Fast, compared to n-gram.
2. Easy to calculate probability.
3. Be able to catch frequent single words.

#### 9. What't the 2-gram models;

Ans: 2-gram language model calculates the sentence probability based on the probability that very 2-adjacent words appears together in a paragraph. The counting is done compared to a collection of all 2-adjacent words.

````python
import random as rd
````

````python
def create_grammar(grammar_str, split='=>', line_split='\n'):
    grammar = {}
    for line in grammar_str.split(line_split):
        if not line.strip(): continue
        exp, stmt = line.split(split)
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar
````

````python
def generate(gram, target):
    if target not in gram: return target #means target is a terminal expression
    
    expand = [generate(gram, t) for t in rd.choice(gram[target])]
    return ''.join([e if e != '/n' else '\n' for e in expand if e != 'null'])
    
````

````python
generate(gram = create_grammar(simple_grammar), target='sentence')
````

TODO: 然后，定义一个函数，generate_n，将generate扩展，使其能够生成n个句子:

````python
def generate_n(gram, target, n):
    result = ''
    for i in range(n):
        result = result + '\n' + generate(gram = create_grammar(gram), target = target)
    print(result)
    return result
````

Code for Question 2:

````python
import random as rd
import pandas as pd
import re
import jieba
from collections import Counter
from functools import reduce
from operator import add, mul
import numpy as np
````

````python
filename = 'train.txt'
content = open(filename, encoding = 'utf-8', errors='ignore').readlines()
````

````python
def token(string):
    return re.sub('[a-zA-Z0-9\s+\.\!\/_,$%^*(+\"\']+|[+——！，。?、~@#￥%……&*（）？-]+','', string)
````

````python
content_clean = [''.join(token(str(a))) for a in content]
#content_clean
#len(content_clean)
````

````python
with open('content_clean.txt','w',errors='ignore') as f: # write the cleaned content into a new file
    for a in content_clean:
        f.write(a + '\n')
````

````python
def cut(string): return list(jieba.cut(string))
````

````python
TOKEN = []
for i, line in enumerate((open('content_clean.txt'))):
    TOKEN += cut(line)

words_count = Counter(TOKEN)
#words_count.most_common(100)
````

````python
def prob_1(word):
    return words_count[word] / len(TOKEN)
````

````python
TOKEN = [str(t) for t in TOKEN]
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-2]))]
words_count_2 = Counter(TOKEN_2_GRAM)
````

````python
def prob_2(word1, word2):
    if word1 + word2 in words_count_2: return words_count_2[word1 + word2] / len(TOKEN_2_GRAM)
    else: return 1 / len(TOKEN_2_GRAM)
````

````python
def get_probability(sentence):
    words = cut(sentence)
    sentence_prob = 1
    for i, word in enumerate(words[:-1]):
        next_word = words[i + 1]
        probability = prob_2(word, next_word)
        sentence_prob *= probability
    return sentence_prob
````

````python
for i, line in enumerate(open('content_clean.txt')):
    print('Sentence {}: {}Probability: {}'.format(i, line, get_probability(line)))
````

能够让list按照第1个元素进行排序, 但是是递减的顺序。

>

````python
def generate_best(gram, target, n):
    result = list()
    content = generate_n(gram, target, n)
    #print('******content is *********', content)
    for line in content.splitlines():
        sentence_prob = get_probability(line)
        result.append((line, sentence_prob))
    #sort by probability
    return sorted(result, key=lambda x: x[1], reverse=True)
````

Q: 这个模型有什么问题？ 你准备如何提升？ 

Ans: 
问题：这个模型无法根据名词搭配其合适的介词、形容词、动词、名词，因而导致很多句子不合逻辑，言不达意。

提升方案：以寻找合适的形容词为例，我们可以通过计算该形容词和某名词在同一个句子中出现的概率，判断该名词和形容词是否搭配。对概率进行排序，概率最高的即为最合理的搭配。将大量含有该名词的句子结合在一起做计算，得出的概率会更准确。这个提升方法还能应用于名词与介词，名词和动词，名词与名词的搭配。从而创造出更合乎逻辑的句子。
